4. **Domanda:** Il valore è vicino a 0 o a 1? Cosa significa?

```r
# 👇 Scrivi qui il tuo codice per l'Esercizio 5
N <- length(cyl)
ni <- table(cyl)
fi <- ni/N
Ni <- cumsum(ni)
Fi <- cumsum(fi)

distr_cyl <- as.data.frame(cbind(ni,fi,Ni,Fi))
distr_cyl

gini.index <- function(x){
    ni <- table(x)
    fi <- ni/length(x)
    g <- 1 - sum(fi^2)
    J <- length(table(x))
    g_norm <- g/(J/J-1)
    return(g); return(g_norm)
}

gini.index(cyl)
#close to 1, means that cyl is well dispersed accross the dataset.
```

```r
# 🟢 SOLUZIONE Esercizio 5

# 1. Tabella frequenze
table(cyl)

# 2. Funzione indice di Gini normalizzato
gini.index <- function(x) {
  ni <- table(x)
  fi <- ni / length(x)
  fi2 <- fi^2
  J <- length(table(x))
  gini <- 1 - sum(fi2)
  gini.norm <- gini / ((J - 1) / J)
  return(gini.norm)
}

# 3. Calcolo
gini.index(cyl)

# 4. Risposta: valore vicino a 1 -> elevata eterogeneità
#    (le tre categorie 4,6,8 sono abbastanza bilanciate)
```

## Esercizio 6 — Simulazione di Probabilità

Immagina di lanciare due dadi a 6 facce, ma se il primo dado è pari allora sommi i valori,
se è dispari moltiplichi i valori.

1. Scrivi una funzione `dado_condizionato(n)` che simula questo esperimento n volte
2. Esegui la simulazione con n = 100.000
3. Crea un istogramma con ggplot2 che mostri la distribuzione dei risultati
4. **Domanda:** Che forma ha la distribuzione?

```r
# 👇 Scrivi qui il tuo codice per l'Esercizio 6
dado_condizionato <- function(n){
    d1 = sequence(1,6,1)
    d2 = sequence(1,6,1)

    t1 = sample(d1, n, replace = TRUE)
    t2 = sample(d1, n, replace = TRUE)

    results <- ifelse(t1%%2 ==0, t1+t2, t1*t2)
    return(results)
}

library(ggplot2)

ggplot() +
    geom_histogram(aes(x = dado_condizionato(10)),
                    stat= "count")

ggplot() +
    geom_histogram(aes(x = dado_condizionato(1000)),
                    stat= "count")

ggplot() +
    geom_histogram(aes(x = dado_condizionato(1000000)),
                    stat= "count")

# they are fragmented

values <- dado_condizionato(10000)
resol = 10
values_cl = cut(values, seq(1,max(values), (max(values)-1)/resol))
fr_val_cl <- as.data.frame(table(values_cl))
fr_val_cl

# no hist cause it wans raw data, not aggregated
ggplot(fr_val_cl) +
    geom_col(aes(x = values_cl, y=Freq),
            col = "red", fill = "orange"
            )
```

```r
# 🟢 SOLUZIONE Esercizio 6

library(ggplot2)

# 1. Funzione di simulazione
dado_condizionato <- function(n) {
  dado <- 1:6
  d1 <- sample(dado, n, replace = TRUE)
  d2 <- sample(dado, n, replace = TRUE)
  
  # Se d1 è pari -> somma, altrimenti -> prodotto
  risultato <- ifelse(d1 %% 2 == 0, d1 + d2, d1 * d2)
  return(risultato)
}

# 2. Simulazione
lanci <- dado_condizionato(100000)

# 3. Istogramma
ggplot() +
  geom_histogram(aes(x = lanci),
                 stat = "count",
                 col = "black",
                 fill = "lightblue") +
  labs(title = "Distribuzione dado condizionato (pari=somma, dispari=prodotto)",
       x = "Risultato", y = "Frequenza")

# 4. Risposta: distribuzione bimodale/multimodale con due "blocchi" distinti
#    (uno per le somme, uno per i prodotti)
```

## Esercizio 7 — Distribuzione Normale

Verifichiamo se `mpg` segue una distribuzione approssimativamente normale.

1. Calcola media e deviazione standard di `mpg`
2. Genera una distribuzione normale con la stessa media e deviazione standard (100.000 valori)
3. Sovrapponi le due densità (quella reale di mpg e quella normale teorica)
4. Standardizza `mpg` (Z = (mpg - mean(mpg)) / sd(mpg)) e verifica che la densità corrisponda a N(0,1)
5. Usa `pnorm()` per calcolare P(mpg ≤ 20) e P(mpg > 25)
6. **Domanda:** Secondo te, mpg è normale? Perché?

```r
# 👇 Scrivi qui il tuo codice per l'Esercizio 7
mu_mpg = mean(mpg)
sig_mpg = sd(mpg)

n_distr <- rnorm(100000, mu_mpg, sig_mpg)



ggplot()+
    geom_density(aes(x = mpg), color = "#570987", fill="purple", alpha = 0.3) +
    geom_density(aes(x = n_distr), color = "darkgreen", fill = "green", alpha = 0.3)
```

```r
# 🟢 SOLUZIONE Esercizio 7

# 1. Media e deviazione standard
mu_mpg <- mean(mpg)
sd_mpg <- sd(mpg)
mu_mpg; sd_mpg

# 2. Genera normale teorica
norm_mpg <- rnorm(100000, mu_mpg, sd_mpg)

# 3. Sovrapposizione densità
plot(density(mpg), col = "blue", lwd = 2,
     main = "Densità di mpg vs Normale teorica",
     xlab = "mpg", ylab = "Densità")
lines(density(norm_mpg), col = "red", lwd = 2, lty = 2)
legend("topright", legend = c("mpg reale", "Normale teorica"),
       col = c("blue", "red"), lwd = 2, lty = c(1, 2))

# 4. Standardizzazione
Z_mpg <- (mpg - mu_mpg) / sd_mpg
plot(density(Z_mpg), col = "blue", lwd = 2,
     main = "mpg standardizzata vs N(0,1)",
     xlab = "Z", ylab = "Densità")
lines(density(rnorm(100000, 0, 1)), col = "red", lty = 2, lwd = 2)
legend("topright", legend = c("mpg std", "N(0,1)"),
       col = c("blue", "red"), lwd = 2, lty = c(1, 2))

# 5. Probabilità con pnorm
pnorm(20, mu_mpg, sd_mpg)  # P(mpg <= 20)
1 - pnorm(25, mu_mpg, sd_mpg)  # P(mpg > 25)

# 6. Risposta: mpg non è perfettamente normale (pochi dati, n=32),
#    ma la forma è approssimativamente simmetrica
```

Analizziamo la forma della distribuzione di `hp` (potenza) e `wt` (peso).

1. Calcola **asimmetria e curtosi** a mano per `hp` (usa le formule dei momenti)
2. Verifica con il pacchetto `moments`
3. Calcola asimmetria e curtosi anche per `wt`
4. Crea un istogramma della densità di `hp` con ggplot2
5. **Domanda:** La distribuzione di hp è simmetrica? Ha code pesanti? Confronta media e mediana.

```r
# 👇 Scrivi qui il tuo codice per l'Esercizio 8
```

```r
# 🟢 SOLUZIONE Esercizio 8

library(ggplot2)
library(moments)

# 1. Asimmetria e curtosi a mano per hp
mu_hp <- mean(hp)
sigma_hp <- sd(hp)
n_hp <- length(hp)

m3_hp <- sum((hp - mu_hp)^3) / n_hp
asim_hp <- m3_hp / sigma_hp^3
asim_hp

m4_hp <- sum((hp - mu_hp)^4) / n_hp
curt_hp <- m4_hp / sigma_hp^4 - 3
curt_hp

# 2. Verifica con moments
skewness(hp)
kurtosis(hp) - 3

# 3. Asimmetria e curtosi per wt
skewness(wt)
kurtosis(wt) - 3

# 4. Istogramma densità hp
ggplot() +
  geom_histogram(aes(x = hp, y = stat(density)),
                 col = "lightblue", fill = "lightblue") +
  labs(title = "Distribuzione di hp", x = "Potenza (hp)", y = "Densità")

# 5. Confronto media e mediana
mean(hp); median(hp)
# media > mediana -> asimmetria positiva (coda a destra)
# curtosi positiva -> code leggermente più pesanti della normale
```
