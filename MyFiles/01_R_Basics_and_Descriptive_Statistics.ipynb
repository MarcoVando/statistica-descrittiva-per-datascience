## 1. R BASICS

### 1.1 Variables & Assignment

In R, variables are assigned using `<-` (preferred) or `=`.  
Use `print()` to display values.

```r
# Assigning variables
a <- 5
b <- "Hello, World!"
c = 56          # = also works, but <- is conventional

print(a)
print(b)
print(c)
```

### 1.2 Data Types

R has several basic data types: `numeric`, `character`, `logical`.  
Use `typeof()` to check the type of an object.

```r
# Basic data types
print(typeof(5))          # numeric (double)
print(typeof("ciao"))    # character
print(typeof(TRUE))       # logical
print(typeof(5L))         # integer (L suffix)
```

```r
# Logical comparisons
print(TRUE)
print(FALSE)
print(5 == 7)   # FALSE
print(5 > 3)    # TRUE
```

### 1.3 Vectors

Vectors are 1D arrays where **all elements must have the same type**.  
Create them with `c()` (combine/concatenate).  
Operations are **element-wise**.

```r
# Creating vectors
x <- c(1, 2, 3, 4, 5)
y <- c(5, 10, 4, 8, 2)

print(x)
print(y)
```

```r
# Element-wise operations
x * y          # (1*5, 2*10, 3*4, 4*8, 5*2)
5 + x          # scalar broadcast: (6,7,8,9,10)
```

```r
# Concatenating vectors
z <- c(x, y)
print(z)       # just a longer vector, NOT a matrix
```

### 1.4 Type Coercion

If you mix types in a vector, R automatically **coerces** them to a common type:  
`logical` → `numeric` → `character`

```r
# Type coercion: all become character
m <- c(1, "hello", TRUE)
print(m)
print(typeof(m))   # character
```

### 1.5 Sequences

Use `seq()` to generate sequences (like Python's `range()`).

```r
# Sequences
seq(1, 10, by = 2)   # 1 3 5 7 9
1:10                  # shorthand: 1 2 3 ... 10
```

### 1.6 Matrices

Matrices are 2D arrays. All elements must have the same type.  
R fills **by column** by default.

```r
# Creating a matrix
M <- matrix(1:6, nrow = 2, ncol = 3)
print(M)
# Layout:
# 1 3 5
# 2 4 6
```

### 1.7 Data Frames

A **data frame** is a table where:
- rows = observations
- columns = variables (can have different types)

Internally, it's a **list of column vectors**.

```r
# Creating a data frame
e <- data.frame(
  name = c("Alice", "Bob", "Charlie"),
  age = c(25, 30, 35),
  is_student = c(TRUE, FALSE, TRUE)
)

print(e)
print(typeof(e))   # list
```

### 1.8 Accessing Elements

Different ways to access elements in R:

```r
# Vector access
x[1]           # first element of x

# Data frame access
e[,1]          # first column as vector
e[1]           # first column as data frame
e[1,1]         # element at row 1, column 1
e["name"]     # column by name (returns data frame)
e$name         # column by name (returns vector)
```

### 1.9 Lists

Lists are the most flexible container — they can hold objects of **any type**, including other lists.

```r
# Creating a list
lista <- list(x, y, e, M)
print(lista)

# Access list elements with [[ ]]
lista[[1]]     # first element (vector x)
```

### 1.10 Functions

Functions are first-class objects in R. They can be assigned to variables and passed as arguments.

```r
# Defining a function
my_function <- function(x) {
  return(x^2)
}

my_function(5)   # 25
```

### 1.11 Packages

R can be extended with external packages.  
Install once with `install.packages()`, then load with `library()`.

```r
# Install (run once)
# install.packages("ggplot2")

# Load in current session
library(ggplot2)
```

Use `?functionname` or `help(functionname)` to open documentation.

```r
# Open documentation
# ?list
# help(list)
```

## 2. LOADING & EXPLORING DATA

### 2.1 Reading a CSV File

Use `read.csv()` to load data.  
The `sep` parameter specifies the delimiter (e.g., `";"` for semicolons).

```r
# Load the tonni (tuna) dataset
dati <- read.csv("tonni.csv", sep = ";")

# Print the entire dataset
print(dati)
```

### 2.2 Exploring the Dataset

Key functions to inspect a dataset:

```r
# Structure: variable names, types, preview
str(dati)

# Dimensions: [rows, columns]
dim(dati)

# Summary statistics
summary(dati)
```

```r
# Number of observations (population size)
N <- dim(dati)[1]
print(N)
```

## 3. FREQUENCY DISTRIBUTIONS

### 3.1 Qualitative Variables

`table()` counts occurrences of each category.

```r
# Frequency table for SESSO (sex)
table(dati["SESSO"])

# Absolute frequencies
freq_ass <- table(dati$SESSO)

# Relative frequencies (divide by N)
freq_rel <- table(dati$SESSO) / N

# Combine into one table
distr_freq_sex <- cbind(freq_ass, freq_rel)
print(distr_freq_sex)
```

```r
# Frequency distribution for LOCALITA (location)
freq_ass <- table(dati$LOCALITa)
freq_rel <- table(dati$LOCALITa) / N
distr_freq_loc <- cbind(freq_ass, freq_rel)
print(distr_freq_loc)
```

### 3.2 Creating Classes for a Quantitative Variable

Use `cut()` to divide a numeric variable into class intervals.

```r
# Check min and max of LUNGHEZZA (length)
print(min(dati$LUNGHEZZA))
print(max(dati$LUNGHEZZA))

# Create classes
dati$LUNGHEZZA_CL <- cut(
  dati$LUNGHEZZA,
  breaks = c(10, 15, 20, 25, 30)
)

print(dati$LUNGHEZZA_CL)
```

### 3.3 Frequency Distribution for Classes

Compute absolute frequencies (`ni`), relative frequencies (`fi`),  
cumulative absolute frequencies (`Ni`), and cumulative relative frequencies (`Fi`).

```r
# Absolute frequencies
ni <- table(dati$LUNGHEZZA_CL)

# Relative frequencies
fi <- table(dati$LUNGHEZZA_CL) / N

# Cumulative absolute frequencies
Ni <- cumsum(ni)

# Cumulative relative frequencies
Fi <- Ni / N

print(ni)
print(fi)
print(Ni)
```

```r
# Build final frequency table
distr_freq_lung_cl <- as.data.frame(
  cbind(ni, fi, Ni, Fi)
)

print(distr_freq_lung_cl)
```

```r
# Contingency table (two variables)
print(table(dati$SESSO, dati$LOCALITa))
```

### 3.4 Exporting Results to CSV

```r
# Save frequency table to CSV
write.csv(
  distr_freq_lung_cl,
  file = "distr_freq_lung_cl.csv"
)
```

```r
# Manual median calculation
n <- length(Petal.Length)

if (n %% 2 == 0) {
  median_val <- (Petal.Length[n/2] + Petal.Length[n/2 + 1]) / 2
} else {
  median_val <- Petal.Length[as.integer(n/2) + 1]
}
print(median_val)

# Compare with built-in
print(median(Petal.Length))
```

Let's create classes for Petal.Length and compute the frequency distribution.

```r
# Create classes for Petal.Length
Petal.Length_cl <- cut(Petal.Length, seq(0, 7.1))
print(table(Petal.Length_cl))
```

```r
# Full frequency table with cumulative frequencies
distr_freq <- as.data.frame(
  cbind(
    ni = table(Petal.Length_cl),
    fi = table(Petal.Length_cl) / length(Petal.Length),
    Ni = cumsum(table(Petal.Length_cl)),
    Fi = cumsum(table(Petal.Length_cl)) / length(Petal.Length)
  )
)
print(distr_freq)
```

## 5. AVERAGES

### 5.1 Arithmetic Mean

```r
# Arithmetic mean
print(mean(Petal.Length))

# Manual calculation
print(sum(Petal.Length) / n)
```

### 5.2 Robustness of Median

The median is **robust** to outliers — the mean is not.

```r
# Add extreme outliers
print(mean(c(Petal.Length, 541, 378)))
print(mean(Petal.Length))   # original mean is much lower

print(median(c(Petal.Length, 541, 378)))
print(median(Petal.Length))  # median barely changes
```

### 5.3 Weighted Mean (for Class Data)

When data is grouped into classes, we use the **weighted mean**  
with class frequencies (`ni`) as weights and class midpoints (`cxi`) as values.

```r
# Add class midpoints
distr_freq$cxi <- seq(0.5, 6.5, 1)
print(distr_freq)

# Weighted mean manually
sum(distr_freq$cxi * distr_freq$ni) / sum(distr_freq$ni)

# Using built-in function
weighted.mean(distr_freq$cxi, distr_freq$ni)
```

### 5.4 Geometric Mean

Useful for **percentage changes / growth rates** (e.g., stock returns).  
Not available in base R — we define it manually.

```r
# Example: cell growth data
cells <- c(1000, 2100, 3000, 5000)

# Install quantmod for percentage changes
# install.packages("quantmod")
library(quantmod)

increments <- quantmod::Delt(cells) * 100
print(increments)
```

```r
# Remove the first NA value
increments <- increments[-1, 1]
print(increments)

# Define geometric mean function
geom.mean <- function(x) {
  return(prod(x)^(1 / length(x)))
}

# Compare geometric vs arithmetic mean
print(geom.mean(increments))
print(mean(increments))
```

### 5.5 Harmonic Mean

Useful for **rates** (e.g., speeds over equal distances).

```r
# Example: speeds over equal distances
speed <- c(100, 80, 40, 90)

# Arithmetic mean (not appropriate for rates)
print(mean(speed))

# Harmonic mean (appropriate for rates)
harmonic.mean <- function(x) {
  return(length(x) / sum(1 / x))
}

print(harmonic.mean(speed))
```
