```r
Petal.Length_cl = cut(Petal.Length, seq(0, 7.1))
```

```r
table(Petal.Length_cl)
```

```r
ditr_freq = as.data.frame(
    cbind(ni = table(Petal.Length_cl),
        fi = table(Petal.Length_cl)/length(Petal.Length),
        Ni = cumsum(table(Petal.Length_cl)),
        Fi = cumsum(table(Petal.Length_cl))/length(Petal.Length)
    )
)
```

```r
ditr_freq
```

Other type of avearegs?

```r
cells = c(1000, 2100, 3000, 5000)
```

```r
# install.packages("quantmod")
increments = quantmod::Delt(cells)*100
increments
```

These are percentages ... so geometric ave is the right one

```r
increments = increments[-1,1]   #to avoid the first NA value
```

geom mean does not exist in base package

```r
geom.mean = function(x) {
  return(prod(x)^(1/length(x)))
}
```

```r
geom.mean(increments)
```

```r
mean(increments)
```

```r
speed = c(100, 80, 40 ,90)
mean(speed)
```

```r
#but it's better to use the harmonic mean, because we are dealing with speeds, 
# and the harmonic mean is more appropriate for rates.

harmonic.mean = function(x) {
  return(length(x)/sum(1/x))    #this is 1/mean(x)
}

harmonic.mean(speed)
```
